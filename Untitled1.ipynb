```python
info = users.Location.str.split(', ')

city = []
state = []
country = []
count_no_state = 0    
count_no_country = 0

for i in range(0,len(info)):
    if info[i][0] == ' ' or info[i][0] == '' or info[i][0]=='n/a' or info[i][0] == ',': 
        city.append('other')
    else:
        city.append(info[i][0].lower())

    if(len(info[i])<2):
        state.append('other')
        country.append('other')
        count_no_state += 1
        count_no_country += 1
    else:
        if info[i][1] == ' ' or info[i][1] == '' or info[i][1]=='n/a' or info[i][1] == ',':  
            state.append('other')
            count_no_state += 1            
        else:
            state.append(info[i][1].lower())
        
        if(len(info[i])<3):
            country.append('other')
            count_no_country += 1
        else:
            if info[i][2] == ''or info[i][1] == ',' or info[i][2] == ' ' or info[i][2] == 'n/a':
                country.append('other')
                count_no_country += 1
            else:
                country.append(info[i][2].lower())
        
users = users.drop('Location',axis=1)

temp = []
for ent in city:
    c = ent.split('/')           
    temp.append(c[0])

df_city = pd.DataFrame(temp,columns=['City'])
df_state = pd.DataFrame(state,columns=['State'])
df_country = pd.DataFrame(country,columns=['Country'])

users = pd.concat([users, df_city], axis=1)
users = pd.concat([users, df_state], axis=1)
users = pd.concat([users, df_country], axis=1)

print(count_no_country)   
print(count_no_state)    
```
